```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
print('Lets play rock paper scissors!')

while True:
    print("What do you choose? 0 for ROCK, 1 for PAPER, and 2 for SCISSORS. To quit enter stop")
    player_choice = input().lower()
    computer_choice = random.randint(0,2)
    if player_choice == 'stop':
        break

    # player chosses rock
    if  player_choice == '0' and computer_choice == 0:
        print(f'Player choice:\n {rock}')
        print(f"Computer choice: \n {rock}")
        print('It is a TIE!')
    elif player_choice == '0' and computer_choice == 1:
        print(f'Player choice:\n {rock}')
        print(f"Computer choice: \n {paper}")
        print('YOU LOSE!')
    elif player_choice == '0' and computer_choice == 2:
        print(f'Player choice:\n {rock}')
        print(f"Computer choice: \n {scissors}")
        print('YOU WIN!!!')

    # player chooses paper
    if  player_choice == '1' and computer_choice == 0:
        print(f'Player choice:\n {paper}')
        print(f"Computer choice: \n {rock}")
        print('YOU WIN!!!')
    elif player_choice == '1' and computer_choice == 1:
        print(f'Player choice:\n {paper}')
        print(f"Computer choice: \n {paper}")
        print('It is a TIE!')
    elif player_choice == '1' and computer_choice == 2:
        print(f'Player choice:\n {paper}')
        print(f"Computer choice: \n {scissors}")
        print('YOU LOSE!')


    # player chooses scissors
    if  player_choice == '2' and computer_choice == 0:
        print(f'Player choice:\n {scissors}')
        print(f"Computer choice: \n {rock}")
        print('You LOSE!!')
    elif player_choice == '2' and computer_choice == 1:
        print(f'Player choice:\n {scissors}')
        print(f"Computer choice: \n {paper}")
        print('YOU WIN!!!')
    elif player_choice == '2' and computer_choice == 2:
        print(f'Player choice:\n {scissors}')
        print(f"Computer choice: \n {scissors}")
        print('It is a TIE!!')
```
